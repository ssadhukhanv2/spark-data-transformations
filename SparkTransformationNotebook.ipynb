```
# PySpark Imports
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
# Defining a schema for our Dataframe
my_schema=StructType([StructField("ID",StringType()),StructField("EventDate",StringType())])

# Defining the data
my_rows=[Row("123","04/05/2020"),Row("124","04/05/2020"),Row("125","04/05/2020"),Row("126","04/05/2020")]

# Creating an RDD with 2 partitions
my_rdd=spark.sparkContext.parallelize(my_rows,2)

# Creating Dataframe from RDD
my_df=spark.createDataFrame(my_rdd,my_schema)
```

```
def to_date_df(df,date_format,column_name):
    return df.withColumn(column_name,to_date(col(column_name),date_format))
```

```
# Info about the Input DataFrame
my_df.printSchema()
my_df.show()


# Creating new DataFrame applying the transformation defined in the function to_date_df
new_df=to_date_df(my_df,"M/d/y","EventDate")

# Info about the Transformed DataFrame
new_df.printSchema()
new_df.show()
```
